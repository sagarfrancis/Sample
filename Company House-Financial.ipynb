```python
import arelle
import lxml
import csv
import pandas as pd
import glob
from arelle import Cntlr
tags= pd.read_csv('C:/Users/stekula/Desktop/Financial Tags.csv')
path2 ='C:/Users/stekula/Desktop/NW Project Submission/Financials'
```

```python
i=0
financial_data=[]
files=[]
mypath ='F:/2016Data'
for (dirpath, dirnames, filenames) in walk(mypath):
    for names in filenames:
        filepath=mypath+'/'+names
        with zipfile.ZipFile(filepath,'r') as g:
            for names in g.namelist():
                if names in a:
                    html_path = filepath+'/'+names
                    xbrl = Cntlr.Cntlr().modelManager.load(html_path)
                    col= ['tags', 'value', 'Id','start time', 'end time', 'unitID']
                    try:
                        data=pd.DataFrame(data=[(fact.concept.qname,
                                                                fact.value, 
                                                                fact.contextID,
                                                                fact.context.startDatetime,
                                                                fact.context.endDatetime,
                                                                fact.unitID) for fact in xbrl.facts], columns=col)


                        data['tags'] = data['tags'].apply(lambda x: str(x).split(':')[1])

                        aa = data['tags']
                        b = tags.Tag
                        l = []
                        for item in aa:
                            for item1 in b:
                                if item == item1:
                                    l.append(item)

                        data1 = data.loc[data['tags'].isin(l)]
                        data2 = tags.loc[tags['Tag'].isin(l)]
                        l=[]
                        data1=data1.merge(data2, left_on='tags', right_on='Tag')
                        Financial=data1[['Label','value', 'Id','start time','end time', 'unitID']]
                        financial_data.append(Financial)
                        i+=1
                        print(i)
                    except AttributeError:
                        financial_data.append(pd.DataFrame({'Error':'Attribute Error'},index=[0]))
```

```python
path ='C:/Users/stekula/Desktop/Financials'
import csv
for i in range(len(financial_data)):
    financial_data[i].to_csv(path + '/fo_'+ str(i) +'.csv')    
```
